```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

## Task 2.a

```python
#NOTE Task 2.a
img1 = cv2.imread('images\image1.jpg')      # read the image
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img1_resized = cv2.resize(img1, (384, 256)) # resize the image to 384 x 256 in columns x rows

fig      = plt.figure(figsize=(8, 4))
subplot  = fig.add_subplot(1,1,1)
subplot.set_xlabel('col')
subplot.set_ylabel('row')
subplot.set_title('image 1 (resized)')
subplot.imshow(img1_resized)
```

## Task 2.b

```python
#NOTE Task 2.b
r, g, b = cv2.split(img1) # split the image into three channels

fig_b     = plt.figure(figsize=(24, 15))
subplot_r = fig_b.add_subplot(1, 3, 1)
subplot_r.set_xlabel('col')
subplot_r.set_ylabel('row')
subplot_r.set_title('image 1 (Red channel)')
subplot_r.imshow(r, cmap = plt.cm.gray)

subplot_g = fig_b.add_subplot(1,3,2)
subplot_g.set_xlabel('col')
subplot_g.set_ylabel('row')
subplot_g.set_title('image 1 (Green channel)')
subplot_g.imshow(g, cmap = plt.cm.gray)

subplot_b = fig_b.add_subplot(1,3,3)
subplot_b.set_xlabel('col')
subplot_b.set_ylabel('row')
subplot_b.set_title('image 1 (Blue channel)')
subplot_b.imshow(b, cmap = plt.cm.gray)
```

## Task 2.c

```python
#NOTE Task 2.c
# calculate the histogram of three channels
hist_r = cv2.calcHist([r], [0], None, [256], [0,256]) 
hist_g = cv2.calcHist([g], [0], None, [256], [0,256])
hist_b = cv2.calcHist([b], [0], None, [256], [0,256])

fig_c      = plt.figure(figsize=(18, 6))
histplot_r = fig_c.add_subplot(1, 3, 1)
histplot_g = fig_c.add_subplot(1, 3, 2)
histplot_b = fig_c.add_subplot(1, 3, 3)

histplot_r.plot(hist_r)
histplot_g.plot(hist_g)
histplot_b.plot(hist_b)

histplot_r.set_title('Histogram for Red channel')
histplot_g.set_title('Histogram for green channel')
histplot_b.set_title('Histogram for blue channel')

subplot_r.set_xlabel('value')
subplot_r.set_ylabel('freq')
subplot_g.set_xlabel('value')
subplot_g.set_ylabel('freq')
subplot_b.set_xlabel('value')
subplot_b.set_ylabel('freq')

plt.show()
```

## Task 2.d

```python
#NOTE Task 2.d
rs_r, rs_g, rs_b = cv2.split(img1_resized)
# do the histogram equlization
equ_rs_r         = cv2.equalizeHist(rs_r)
equ_rs_g         = cv2.equalizeHist(rs_g)
equ_rs_b         = cv2.equalizeHist(rs_b)
equ_img1_rs      = cv2.merge([equ_rs_r, equ_rs_g, equ_rs_b]) # merge the three seperate images
equ_img1_rs      = cv2.cvtColor(equ_img1_rs, cv2.COLOR_RGB2GRAY)
hist_equ_img1_rs = cv2.calcHist([equ_img1_rs], [0], None, [256], [0,256])
# do the histogram equlization
equ_r = cv2.equalizeHist(r)
equ_g = cv2.equalizeHist(g)
equ_b = cv2.equalizeHist(b)
# calculate the histogram of three channels
hist_equ_r = cv2.calcHist([equ_r], [0], None, [256], [0,256])
hist_equ_g = cv2.calcHist([equ_g], [0], None, [256], [0,256])
hist_equ_b = cv2.calcHist([equ_b], [0], None, [256], [0,256])

fig_d       = plt.figure(figsize=(8, 32))

equplot_rs  = fig_d.add_subplot(4, 1, 1)
equplot_rs.plot(hist_equ_img1_rs)
equplot_rs.set_title('Histogram for resized image')
equplot_rs.set_xlabel('value')
equplot_rs.set_ylabel('freq')

equplot_r  = fig_d.add_subplot(4, 1, 2)
equplot_r.plot(hist_equ_r)
equplot_r.set_title('Histogram for Red channel')
equplot_r.set_xlabel('value')
equplot_r.set_ylabel('freq')

equplot_g  = fig_d.add_subplot(4, 1, 3)
equplot_g.plot(hist_equ_g)
equplot_g.set_title('Histogram for green channel')
equplot_g.set_xlabel('value')
equplot_g.set_ylabel('freq')

equplot_b  = fig_d.add_subplot(4, 1, 4)
equplot_b.plot(hist_equ_b)
equplot_b.set_title('Histogram for blue channel')
equplot_b.set_xlabel('value')
equplot_b.set_ylabel('freq')

plt.show()
```
